```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

basic = pd.read_csv('data/basic_performance_test.totals.csv')
atomic = pd.read_csv('data/atomic_performance_test.totals.csv')
```

```python
# Add new column for operations per second for each test
basic['Ops per ms'] = basic['Total responses'] / basic['Total time (ms)']
atomic['Ops per ms'] = atomic['Total responses'] / atomic['Total time (ms)']
basic['Ops per second'] = basic['Ops per ms'] * 1000
atomic['Ops per second'] = atomic['Ops per ms'] * 1000


# Compare the two as a bar chart with one bar for basic and one for atomic
fig, ax = plt.subplots()
ax.bar('Basic', basic['Ops per second'], label='Basic')
ax.bar('Atomic', atomic['Ops per second'], label='Atomic')
ax.set_ylabel('Ops per second')
ax.legend()

plt.show()
```
